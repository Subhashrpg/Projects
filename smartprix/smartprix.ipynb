```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
with open(r"D:\Projects\smartprix\smartprix.html","r",encoding="utf-8") as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html,"lxml")
```

```python
for text in soup.find_all("div", class_= "sm-product has-tag has-features has-actions"):
    try:
        names.append(text.find("h2").text)
    except:
        names.append(np.nan)
    try:
        prices.append(text.find("span", class_= "price").text)        
    except:
        prices.append(np.nan)        
    try:
        rating.append(text.find("div", class_= "score rank-2-bg").find("b").text)
    except:
        rating.append(np.nan)
    x = text.find_all("li", class_= "")
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)        
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)     
```

```python
names = []
prices = []
rating = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []
```

```python
len(names)
```

```python
df = pd.DataFrame(
{
    "model": names,
    "price": prices,
    "rating": rating,
    "sim": sim,
    "processor": processor,
    "ram": ram,
    "battery": battery,
    "display": display,
    "camera": camera,
    "card": card,
    "os": os
})
```
